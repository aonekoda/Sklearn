1. 아래의 코드를 수행하여 그래프로 표시할 데이터를 만드시오.   
subplot을 활용하여 아래와 같은 결과를 그래프로 그리시오.

```
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

# TODO

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
````

2. 다음과 같이 sine 그래프와 cosine 그래프를 하나의 그래프로 표시하시오.  
선의 모양, 색깔을 다르게 하고 legend를 표시하시오.

````python
x = np.linspace(0,10,100)

# TODO
sine = np.sin(x)
cosine = np.cos(x)
plt.plot(x, sine, 'r-',label='sine')
plt.plot(x, cosine, 'g--', label='cosine')
plt.legend(loc='upper right')
````
